```python
#OUTLIER DETECTION IN TRANSACTION AMOUNT USING IQR METHOD
Q1 = df['amt'].quantile(0.25)
Q3 = df['amt'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df[(df['amt'] < lower_bound) | (df['amt'] > upper_bound)]
print("Number of outliers in 'amt':", outliers.shape[0])
```

OBSERVATION:Outliers in the 'amt' column were detected using the Interquartile Range (IQR) method. Transactions                 falling below the lower bound (Q1 - 1.5 * IQR) or above the upper bound (Q3 + 1.5 * IQR) were                       identified as outliers. A total of X outliers were found, where X is the number printed by                          outliers.shape[0].

```python
#SKEWNESS ANALYSIS OF TRANSACTION AMOUNT (AMT)
from scipy.stats import skew

# Example for 'amt' column (transaction amount)
skewness_value = skew(df['amt'])
print("Skewness of amt:", f"{skewness_value:.2f}")
```

OBSERVATION:The amt column has a skewness value of 9.85, indicating a highly right-skewed distribution. This suggests that most transaction amounts are relatively small, with a few very large transactions stretching the distribution to the right. This kind of skewness is common in financial data and may require log transformation or scaling to normalize before applying machine learning models.

```python
#LOG TRANSFORMATION OF TRANSACTION AMOUNT TO REDUCE SKEWNESS
import numpy as np
df['amt_log'] = np.log1p(df['amt'])  # log(1 + amt) to handle zeros
print(df)
```

OBSERVATION:This technique compresses large values more than small ones, reducing the impact of extreme                         transactions and making the distribution more symmetric.

```python
#COMPARATIVE DISTRIBUTION OF ORIGINAL VS LOG-TRANSFORMED TRANSACTION AMOUNT 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
sns.histplot(df['amt'], kde=True)
plt.title("Original 'amt' Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['amt_log'], kde=True, color='green')
plt.title("Log-Transformed 'amt_log' Distribution")

plt.tight_layout()
plt.show()
```

OBSERVATION:
The histogram on the left displays the original distribution of the amt (transaction amount) variable, which is highly right-skewed, indicating the presence of numerous small-value transactions and a few large-value outliers.
The histogram on the right shows the distribution after applying a logarithmic transformation (log1p), resulting in the amt_log column. This transformation:
     Compresses the scale of large values,
     Reduces skewness,
     Produces a more symmetrical and normalized distribution.

```python
# COUNT OF FRAUDULENT VS. NON-FRAUDULENT TRANSACTIONS
fraud_counts = df['is_fraud'].value_counts()
print("Fraudulent vs. Non-Fraudulent Transactions:\n", fraud_counts)
```

```python
# CALCULATE FRAUD RATE
fraud_count = df['is_fraud'].sum()  # since fraud is marked as 1
total_count = df.shape[0]
fraud_rate = (fraud_count / total_count) * 100

print(f"Fraud Rate: {fraud_rate:.2f}%")
```

Only 0.24% of the total transactions are marked as fraudulent.
```python
# AVERAGE AND MEDIAN TRANSACTION AMOUNT FOR FRAUD VS. NON-FRAUD
fraud_stats = df.groupby('is_fraud')['amt'].agg(mean_amt=lambda x: round(x.mean(), 2),
    median_amt=lambda x: round(x.median(), 2))
print(fraud_stats)
```

```python
#FRAUD DISTRIBUTION BY GEOGRAPHIC ZONE (NORTH/SOUTH/EAST/WEST)
def region_classification(row):
    if row['lat'] > 37:
        return 'NORTH'
    elif row['lat'] < 30:
        return 'SOUTH'
    elif row['long'] < -100:
        return 'WEST'
    else:
        return 'EAST'

df['region'] = df.apply(region_classification, axis=1)

print(df.groupby(['region', 'is_fraud']).size().unstack())
```

OBSERVATION:
The southern region has the highest fraud rate at 1.74%, despite a smaller number of total transactions. this             suggests that south-based transactions may be riskier per capita. in contrast, the northern region has the                lowest fraud rate even though it has the most transactions overall.

```python
#VISUAL ANALYSIS OF FRAUD DETECTION DATASET 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Log transformation
df['amt_log'] = np.log1p(df['amt'])
df['is_fraud_label'] = df['is_fraud'].map({0: 'Not Fraud', 1: 'Fraud'})

# Set up subplot grid
fig, axes = plt.subplots(3, 2, figsize=(8, 6))
fig.suptitle("Visual Analysis of Fraud Detection Dataset", fontsize=16)

# Histogram of original 'amt'
sns.histplot(df['amt'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Histogram of Transaction Amount (amt)")

# Histogram of log-transformed 'amt'
sns.histplot(df['amt_log'], kde=True, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title("Histogram of Log-Transformed amt_log")

# Bar plot of fraud vs not fraud
sns.countplot(data=df, x='is_fraud_label', hue='is_fraud_label', palette='Set2', ax=axes[1, 0], legend=False)
axes[1, 0].set_title("Bar Plot of Fraud vs Not Fraud")

# Pie chart
fraud_counts = df['is_fraud_label'].value_counts()
axes[1, 1].pie(fraud_counts, labels=fraud_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
axes[1, 1].set_title("Pie Chart of Fraud Distribution")

# Violin plot
sns.violinplot(data=df, x='is_fraud_label', y='amt', hue='is_fraud_label', palette='pastel', ax=axes[2, 0], legend=False)
axes[2, 0].set_title("Violin Plot of amt by Fraud Status")

# Stri plot (sampled for performance)
sample_df = df.sample(n=500, random_state=42)
sns.stripplot(data=sample_df, x='is_fraud_label', y='amt', hue='is_fraud_label',
              palette='muted', dodge=True, alpha=0.6, ax=axes[2, 1])
axes[2, 1].set_title("Strip Plot of amt by Fraud Status")
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

Observation:
The distribution of transaction amounts is highly right-skewed, with most values concentrated in the lower range (under ₹200).

2. HISTOGRAM OF LOG-TRANSFORMED AMOUNT (AMT_LOG)
Observation:
After applying a log transformation, the distribution becomes more symmetrical and normalized.
This improves the scalability and performance of machine learning algorithms and highlights hidden patterns in transaction behavior.
3. BAR PLOT OF FRAUD VS NOT FRAUD

The data is highly imbalanced, with far fewer fraudulent transactions than non-fraudulent ones.

Observation:
Both fraudulent and non-fraudulent transactions have dense low-value distributions.
Fraudulent transactions exhibit a wider spread and longer tails, suggesting they occasionally involve higher amounts, making amount a relevant feature for classification.
